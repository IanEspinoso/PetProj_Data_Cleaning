```python
# in 'Type' we have a 'boat' issue. let's change all those issues into 'boating', and the nans as 'Questionable'

sum(sharks['Type'].isna())
sharks['Type'].fillna('Questionable', inplace=True)
sharks['Type'].mask(sharks['Type'].str.startswith('Boat'),'Boating', inplace=True)
```

```python
# in 'Fatal' we have need to cook everything down to just 'UNKNOWN' (depending on 'nan' ammount), 'Y' and 'N'

sum(sharks['Fatal (Y/N)'].isna())
sharks['Fatal (Y/N)'].fillna('UNKNOWN', inplace=True)
sharks['Fatal (Y/N)'] = sharks['Fatal (Y/N)'].str.upper()
sharks['Fatal (Y/N)'] = sharks['Fatal (Y/N)'].str.strip()

# there are still those SINGLE cases with a 2017 and a M. ok...

sharks['Fatal (Y/N)'] = sharks['Fatal (Y/N)'].replace({'2017':'UNKNOWN','M':'N'})
sharks['Fatal (Y/N)'].unique()
```

```python
# in 'Country', the issues are mainly casing and not being an actual country
# that second part might be tricky, but we might try sending the misfits into 'Area' cells

sum(sharks['Country'].isna())
sharks['Country'].fillna('UNKNOWN', inplace=True)
sharks['Country'] = sharks['Country'].apply(lambda x: x.title())
sharks['Country'] = sharks['Country'].replace('Usa','USA')

# I imported a reliable library called 'pycountry' to check the existence of said countries in the dataframe
```

```python
# defining a search function

def country_search(x):
    """
    Returns the oficial name of the country, as long as a fit is found.
    That means the function pretty much just spell checks.
    Looking forward to implement a fuzzy search.
    """
    from pycountry import countries
    try:
        return countries.get(name=x).name
    except:
        return 'UNKNOWN'
```

```python
# implementing 'Country' into 'Location' if it corresponds to a null cell

sharks['Location'][sharks['Location'].isna()] = sharks['Country'][sharks['Location'].isna()]
```

```python
# remaking column 'Country' with the correct spellings or 'UNKNOWN'

sharks['Country'].apply(country_search)

# that would be disastrous. let's check what are those many unknowns...those are 'USA'. let's correct it.
# fuzzy-searching those countries, we can see that the spelling there is United States and Viet Nam

countries.search_fuzzy('USA')

# let's make that substitution

sharks['Country'].replace({'USA':'United States','Vietnam':'Viet Nam'}, inplace=True)
```

```python
sharks['Country'] = sharks['Country'].apply(country_search)
```

```python
# now, let's find out those cases where 'location' is just a repetition of 'country', and take it out

sharks['Location'][sharks['Location']==sharks['Country']] = 'UNKNOWN'
```
